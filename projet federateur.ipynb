```
from nltk.corpus import stopwords
import nltk

def segmenter_texte(texte):
  tokenizer = nltk.tokenize.TreebankWordTokenizer()
  tokens= tokenizer.tokenize(texte)
  tokens_mini=[]
  for w in tokens:
    tokens_mini.append(w.lower())
    #  nltk.download('stopwords')
  for w in tokens_mini[:]:
    if w in stopwords.words('english') :
      tokens_mini.remove(w)
  tokens=tokens_mini
  stemmer = nltk.stem.PorterStemmer()
  ch=" ".join(stemmer.stem(token) for token in tokens)
  return ch.split()
```

# Etape 4 : Modèle d’apprentissage automatique (Modeling)


```
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
import numpy as np
```

```
from nltk.corpus import stopwords
import nltk
from sklearn.feature_extraction.text import CountVectorizer


```

```
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from sklearn.feature_extraction.text import TfidfTransformer


```

```
from sklearn import svm
```

```
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
import numpy as np
```

```
pipeline1 = [("vect", CountVectorizer()), 
        ("tfidf", TfidfTransformer()),
        ("svm", SVC())]
pipe = Pipeline(pipeline1)
pipe.named_steps
kernel="rbf"
```

```

params = {
    'vect__min_df': np.linspace(0.005, 0.05, 5),
    'vect__ngram_range': ((1, 1),(1, 2)), 
    'vect__decode_error': ('ignore','UTF-8'),
    'tfidf__use_idf': (True, False),
    'svm__C':  [0.1, 1, 10],
    # 'svm__gamma': [0.1, 1, 10],
    # 'svm__kernel': [0.1, 1, 10]
}
```

```

gs1=GridSearchCV(pipe,param_grid=params,cv=5,n_jobs=-1)
gs1=gs1.fit(X_train, y_train)
gs1.best_params_
```

```
from sklearn.naive_bayes import MultinomialNB
pipeline2 = [("vect", CountVectorizer()), 
        ("tfidf", TfidfTransformer()),
        ("multinomialNB", MultinomialNB(alpha = 1.0))]
pipe = Pipeline(pipeline2)
pipe.named_steps
```

```
params = {
    'vect__min_df': np.linspace(0.005, 0.05, 5),
    'vect__ngram_range': ((1, 1),(1, 2)), 
    'vect__decode_error': ('ignore','UTF-8'),
    'tfidf__use_idf': (True, False),
    'multinomialNB__fit_prior' : (False,True)
}
```

```
gs2=GridSearchCV(pipe,param_grid=params,cv=5,n_jobs=-1)
gs2=gs2.fit(X_train, y_train)
gs2.best_params_
```

# Etape 5 : Evaluation (Evaluation)

```
y_test_predicted1=gs1.predict(X_test)
y_test_predicted1
```

```
y_test_predicted2=gs2.predict(X_test)
y_test_predicted2
```

```
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_test,y_test_predicted1))
print(confusion_matrix(y_test,y_test_predicted1))
```

```
print(accuracy_score(y_test,y_test_predicted2))
print(confusion_matrix(y_test,y_test_predicted2))
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_test_predicted1))
```

```
print(classification_report(y_test,y_test_predicted2))
```
