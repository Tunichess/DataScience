````
from sklearn.base import BaseEstimator, TransformerMixin
import datetime
````

````
class Transformee(BaseEstimator, TransformerMixin):
    
    def __init__(self, surface_moyenne=False, date_to="annee"):
        self.surface_moyenne=surface_moyenne
        self.date_to=date_to
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        X_=X.copy()
        
        # Traitement sur les donn√©es (X_)
        # Etape 1. Suppression des valeurs NaN
        X_.fillna({'surface': X_.surface.mean() , 
                   'nb_chambre': X_.nb_chambre.mean()},
                   inplace=True)
        # Etape 2. Convertir la date de creation annee ou age
        X_.date_creation=pd.to_datetime(X_.date_creation)
        if self.date_to=='annee':
            X_['annee']=X_['date_creation'].apply(lambda date : date.year)
        elif self.date_to=='age':
            X_['age']=X_['date_creation'].apply(lambda date : 
                                                    datetime.datetime.today().year-date.year)
        X_.drop('date_creation', axis=1, inplace=True, errors='ignore')
        # Etape 3. Convertir ou pas la surface et nb_chambre en surface moyenne
        if self.surface_moyenne==True:
            X_['surface_moyenne']=X_.apply(lambda row : row['surface']/row['nb_chambre'], axis=1)
            X_.drop(['surface','nb_chambre'], axis=1, inplace=True, errors='ignore')
        # Etape 4. Convertir la couleur en codes binaires avec Hot-Encoding
        df_couleur_encoded=pd.get_dummies(X_.couleur, prefix='couleur')
        X_=pd.concat([X_,df_couleur_encoded],axis=1)
        X_.drop('couleur', axis=1, inplace=True, errors='ignore')
        return X_
````
