```
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```
te =TransactionEncoder()
```

```
dataset=[['Pain','Lait'],['Pain','Couche','Biere','Oeuf'],['Lait','Couche','Biere','Coca'],['Pain','Lait','Couche','Biere'],['Pain','Lait','Couche','Coca']]
```

```
te.fit(dataset)
```

```
te.columns_
```

```
te_ary=te.transform(dataset)
```

```
type(te_ary)
```

```
te_ary
```

```
df=pd.DataFrame(te_ary, columns=te.columns_)
```

```
df
```

```
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```
df_itemsets=apriori(df, min_support=0.6)
```

```
df_itemsets
```

```
df_itemsets=apriori(df, min_support=0.4,use_colnames=True)
```

```
df_itemsets
```

```
df_itemsets=apriori(df, min_support=0.8,use_colnames=True)
```

```
df_itemsets
```

```
df_itemsets=apriori(df,min_support=0.6,use_colnames=True)
df_itemsets['length']=df_itemsets['itemsets'].apply(lambda x: len(x))
```

```
df_itemsets
```

```
df_itemsets[(df_itemsets['length']==2) & (df_itemsets['support']>=0.6)]
```

```
df_itemsets[df_itemsets['itemsets']=={'Couche','Biere'}] 
```

```
rules=association_rules(df_itemsets,metric="lift",min_threshold=1)
rules
```

```
rules =association_rules(df_itemsets,metric="confidence",min_threshold=1)
rules
```
