#Regression Deaths






```
import pandas as pd
import numpy as np
```

```
data_url="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
```

```
deaths_df=pd.read_csv(data_url)
```

```
deaths_df.shape
```

```
deaths_df.head()
```

```
deaths_tunisia=deaths_df.loc[deaths_df['Country/Region']=='Tunisia',deaths_df.columns[4:]]
```

```
deaths_tunisia
```

```
dates= deaths_tunisia.columns
```

```
numero_dates=np.arange(0,len(dates),1)
```

```
deaths_nb=deaths_tunisia.to_numpy().reshape(-1)
```

```
dates
```

```
numero_dates
```

```
nombre=deaths_nb
nombre
```

```
df_tunisia = pd.DataFrame({"num_date": numero_dates, "nombre": nombre} , index=dates)
df_tunisia
```

```
X=df_tunisia[['num_date']]
X
```

```
y=df_tunisia.nombre
y
```

```
import matplotlib.pyplot as plt
plt.plot(X,y,'go')
plt.plot([0,130],[5,5],'b--')
plt.plot([0,130],[45,45],'b--')
plt.xlabel('temps')
plt.ylabel('Cas mort')
```

```
bins=[0,5,45,100]
```

```
import numpy as np
y_binned=np.digitize(y,bins)
y_binned
```

```
from sklearn.model_selection import train_test_split

```

```
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3, stratify=y_binned)
```

```
plt.plot(X_train,y_train,'go')
```

```
plt.plot(X_test,y_test,'go')
```

##Modele Lineaire:

```
from sklearn.linear_model import LinearRegression
```

```
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```
y_test_predicted=lr.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error
```

```
mean_squared_error(y_test,y_test_predicted)
```

```
from sklearn.metrics import r2_score
r2_score(y_test,y_test_predicted)
```

```
plt.plot(X_train,y_train,'go',label='train data')
plt.plot(X,lr.predict(X),'r-',label='modele lineaire')
plt.xlabel('temps')
plt.ylabel('nombre de cas mort')
plt.legend()
```

##Modele Logistique:

```
def logistic_model(x,a,b,c):
  return c/(1+np.exp(-(x-b)/a))
```

```
X_train=X_train.to_numpy().reshape(-1)
y_train=y_train.to_numpy().reshape(-1)
```

```
from scipy.optimize import curve_fit
```

```
params = curve_fit(logistic_model,X_train,y_train)
```

```
a=params[0][0]
b=params[0][1]
c=params[0][2]
```

```
y_test_log_pred=log_pred=logistic_model(X_test,a,b,c)
```

```
mean_squared_error(y_test,y_test_log_pred)
```

```
r2_score(y_test,y_test_log_pred)
```

```
y_log_pred=logistic_model(X,a,b,c)
plt.plot(X_train,y_train,'go',label='train data')
plt.plot(X,y_log_pred,'r-',label='modele logistique')
plt.xlabel('temps')
plt.ylabel('nombre de cas mort')
plt.legend()
```
