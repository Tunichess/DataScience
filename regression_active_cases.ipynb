## **Regression Active Cases**

```
import pandas as pd
import numpy as np
```

```
data_url1="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
```

```
data_url2="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
```

```
data_url3="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
```

```
confirmed_df=pd.read_csv(data_url1)
recovered_df=pd.read_csv(data_url2)
deaths_df=pd.read_csv(data_url3)
```

```
confirmed_df.shape
```

```
recovered_df.shape
```

```
deaths_df.shape
```

```
confirmed_df.head()
recovered_df.head()
deaths_df.head()
```

```
confirmed_df['Country/Region'][-6:]
```

```
confirmed_tunisia=confirmed_df.loc[confirmed_df['Country/Region']=='Tunisia',confirmed_df.columns[4:]]
recovered_tunisia=recovered_df.loc[recovered_df['Country/Region']=='Tunisia',recovered_df.columns[4:]]
deaths_tunisia=deaths_df.loc[deaths_df['Country/Region']=='Tunisia',deaths_df.columns[4:]]
```

```
confirmed_tunisia
recovered_tunisia
deaths_tunisia
```

```
dates= confirmed_tunisia.columns
```

```
dates.shape
```

```
numero_dates=np.arange(0,len(dates),1)
```

```
confirmed_nb=confirmed_tunisia.to_numpy().reshape(-1)
recovered_nb=recovered_tunisia.to_numpy().reshape(-1)
deaths_nb=deaths_tunisia.to_numpy().reshape(-1)
```

```
confirmed_nb
```

```
dates
```

```
numero_dates
```

```
active_cases_nb=confirmed_nb-(recovered_nb+deaths_nb)
active_cases_nb
```

```
nombre=active_cases_nb
nombre
```

```
df_tunisia = pd.DataFrame({"num_date": numero_dates, "nombre":active_cases_nb } , index=dates)
df_tunisia
```

```
X=df_tunisia[['num_date']]
X
```

```
y=df_tunisia.nombre
y
```

```
import matplotlib.pyplot as plt
plt.plot(X,y,'go')
# plt.plot([0,130],[100,100],'b--')
# plt.plot([0,130],[400,400],'b--')
plt.xlabel('temps')
plt.ylabel('Cas active')
```

```
bins=[0,100,400,1200]
```

```
import numpy as np
y_binned=np.digitize(y,bins)
y_binned
```

```
from sklearn.model_selection import train_test_split

```

```
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3, stratify=y_binned)
```

```
plt.plot(X_train,y_train,'go')
```

```
plt.plot(X_test,y_test,'go')
```

##Modele Lineaire




```
from sklearn.linear_model import LinearRegression
```

```
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```
y_test_predicted=lr.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error
```

```
mean_squared_error(y_test,y_test_predicted)
```

```
from sklearn.metrics import r2_score
r2_score(y_test,y_test_predicted)
```

```
plt.plot(X_train,y_train,'go',label='train data')
plt.plot(X,lr.predict(X),'r-',label='modele lineaire')
plt.xlabel('temps')
plt.ylabel('nombre de cas confirmés')
plt.legend()
```

##Modele Logistique




```
def logistic_model(x,a,b,c):
  return c/(1+np.exp(-(x-b)/a))
```

```
X_train=X_train.to_numpy().reshape(-1)
y_train=y_train.to_numpy().reshape(-1)
```

```
from scipy.optimize import curve_fit
```

```
params = curve_fit(logistic_model,X_train,y_train)
```

```
a=params[0][0]
b=params[0][1]
c=params[0][2]
```

```
y_test_log_pred=log_pred=logistic_model(X_test,a,b,c)
```

```
mean_squared_error(y_test,y_test_log_pred)
```

```
r2_score(y_test,y_test_log_pred)
```

```
y_log_pred=logistic_model(X,a,b,c)
plt.plot(X_train,y_train,'go',label='train data')
plt.plot(X,y_log_pred,'r-',label='modele logistique')
plt.xlabel('temps')
plt.ylabel('nombre des cas actives')
plt.legend()
```

##Modele Polynomial

```
from sklearn.preprocessing import PolynomialFeatures
```

```
poly_reg = PolynomialFeatures(degree=14)
```

```
X_poly = poly_reg.fit_transform(X_train.reshape(-1,1))
```

```
pol_reg = LinearRegression()
```

```
pol_reg.fit(X_poly, y_train)
```

```
y_test_pred=pol_reg.predict(poly_reg.fit_transform(X_test))
```

```
r2_score(y_test,y_test_pred)
```

```
plt.plot(X_train,y_train,'go',label='train data')
plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)),'r-',label='modele Polynomial')
plt.xlabel('temps')
plt.ylabel('nombre des cas actives')
plt.legend()
```

##Modele de Gauss




```
from scipy.stats import norm
from scipy import asarray as ar,exp
def gauss(x,a,b,c):
  mu=b
  sig=a
  return a*exp(-(x-b)**2/(2*c**2))
```

```
from scipy.optimize import curve_fit
params = curve_fit(gauss,X_train,y_train,p0=[20,20,20])
```

```
a=params[0][0]
b=params[0][1]
c=params[0][2]
```

```
y_test_log_pred=log_pred=gauss(X_test,a,b,c)
```

```
mean_squared_error(y_test,y_test_log_pred)
```

```
r2_score(y_test,y_test_log_pred)
```

```
y_log_pred=gauss(X,a,b,c)
plt.plot(X_train,y_train,'go',label='train data')
plt.plot(X,y_log_pred,'r-',label='modele gauss')
plt.xlabel('temps')
plt.ylabel('nombre des cas actives')
plt.legend()
```
